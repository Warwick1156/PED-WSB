```python
import pandas as pd
import ast

from os import path
from glob import glob

data_dir = path.join('..', 'data')
dataset_dir = path.join(data_dir, 'datasets')

file_list = glob(path.join(dataset_dir, '*.csv'))
dataset_file = max(file_list, key=path.getctime)
print(dataset_file)

data = data = pd.read_csv(path.join(data_dir, dataset_file), sep='`')
```

```python
test = data.imgs_colors[data.imgs_colors != '[]']                       
```

```python
test = [ast.literal_eval(x) for x in test]
```

```python
test
```

```python
def proces_hsv(colours: list):
    h, s, v = [], [], []
    for colour in colours:
        colour = colour[0]
        h.append(colour[0])
        s.append(colour[1])
        v.append(colour[2])
        
    return h, s, v

h, s, v = proces_hsv(test)
```

```python
print(min(h), max(h))
print(min(s), max(s))
print(min(v), max(v))
```
