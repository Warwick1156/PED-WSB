```python
def get_cumsum(values):
    
    if len(values) == 0:
        return []
    
    acc = values[0]
    result = [acc]
    for i,e in enumerate(values[1:]):
        acc += e
        result.extend([acc])
        
    return result
```

```python
from tqdm.notebook import tqdm
```

```python
idx = list(set(data.post_id))

import pandas as pd
import numpy as np

from multiprocessing import cpu_count, Pool

def run_run_run(idx):
    i = 0
    for post_id in tqdm(idx):
        i += 1
        mask = data.post_id == post_id
        c = get_cumsum(data.new_comments.loc[mask].tolist())
        data.cumulative.loc[mask] = c
        
        if i%25==0:
            print(i)
    
idx_split = np.array_split(idx, 8)
pool = Pool(8)
pool.map(run_run_run, idx_split)
pool.close()
pool.join()
```
